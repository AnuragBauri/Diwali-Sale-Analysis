### Exploratory Data Analsis

### Gender

```python
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen= df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Gender',y='Amount',data=sales_gen,hue='Gender',palette={'green','red'},legend=False)
plt.xlabel('Gender')
plt.ylabel('Amount')
plt.title('Total Amount of Gender')
plt.show()
```

#### From above graphs we can see that most of the buyers are females greater than male

### Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

#### From above graphs we can see that most of the buyers are age between 26-35 yrs

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age,hue='Age Group',palette='Set1')
```

### State

```python
# Total Numbers of orders from Top 10 States

sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(sales_state,x='State',y='Orders',hue='State')
```

### Occupation


```python
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(10,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Create the grouped dataset
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Plot with unique colors
sns.barplot(
    data=sales_state,
    x='Occupation',
    y='Amount',
    hue='Occupation',        # tell seaborn to color by Occupation
    dodge=False,
    legend=False,            # remove duplicate legend
    palette='Set2'           # use a built-in palette OR define custom list
)
```

## From above Graphs we can see that most of the buyers are working in IT Sector
